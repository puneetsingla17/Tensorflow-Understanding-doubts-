```python
a=np.random.randint(0,10,size=[5,4])
b=np.random.randint(0,10,size=[5,4])
c=np.random.randint(0,10,size=[5,4])
d=np.random.randint(0,10,size=[5,4])
```

```python
tf.stack([a,b,c,d],axis=-1).eval().shape   #very important stacking is only possible when we have all tensors of same dims
```
